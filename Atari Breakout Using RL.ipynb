```python
env_name = "Breakout-v0"
env = gym.make(env_name)
```

```python
episode = 5
for episode in range(1,episode+1):
    state = env.reset()
    done = False
    score = 0
    
    while not done:
        env.render()
        action = env.action_space.sample()
        n_state, reward, done, info = env.step(action)
        score += reward
    print('Episode: {} Score: {}'.format(episode,score))
#env.close()
```

## Vectorise Environment and Train Model

```python
# This code makes 4 different environment for breakout, to train 4 env at the same time to speed up training
env = make_atari_env('Breakout-v0', n_envs = 4, seed = 0)
env = VecFrameStack(env, n_stack=4)
```

```python
log_path = os.path.join('Training', 'Logs')
model = A2C('CnnPolicy', env, verbose=1, tensorboard_log = log_path)
```

```python
a2c_path = os.path.join('Training', 'Saved_Models', 'A2C_1M_Breakout')
```
